```python
import folium
```

```python
data = [
    ("47.8030008","13.0435397"),
    ("47.8233907","13.0400552")
]
data1 = [
    [47.8030008,13.0435397],
    [47.8233907,13.0400552]
]
    
    
```

```python
m=folium.Map(location=[47.8053133,13.0410201],zoom_start=13)
m
```

```python
for cords in data:
    folium.Marker(location=[cords[0], cords[1]]).add_to(m)
m
```

```python
names = [
    "Schnaitl",
    "Techno Z"
]
```

```python
for cords, name in zip(data, names):
    folium.Marker(location=[cords[0], cords[1]],
                  popup=f"Name:<br>{name}"
                  ).add_to(m)
m
```

```python
line=folium.PolyLine(locations=data1,weight=5)
m.add_child(line)
```
